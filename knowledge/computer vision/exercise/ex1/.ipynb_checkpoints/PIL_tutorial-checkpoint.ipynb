## Convert file to jpeg

直接文件名后加扩展名就可以自动转换

```python
import os, sys

infile = 'lena.png'
f, e = os.path.splitext(infile)
print ('name:', f,'\n','format:', e)

outfile = f + '.jpg'
if infile != outfile:
    try:
        Image.open(infile).save(outfile)
    except:
        print('cannot convert file:', outfile)
```

# thumbnail

仅仅是缩小了图片大小而已

`im.thumbnail(size)`

```python
import os, sys

infile = 'lena.png'
size = (128, 128)

f, _ = os.path.splitext(infile)
outfile = f + '.thumbnail'
if infile != outfile:
    try:
        im = Image.open(infile)
        im.thumbnail(size)
        im.save(outfile, 'JPEG')
    except IOError:
        print("cannot create thumbnail for", infile)
```

# cutting, pasting and merging images

## Copying a subrectangle from an image

- (0, 0, 400, 400) -> (left, upper, right, lower)
- (0, 0) is the __upper left__ corner

```python
box = (0, 0, 400, 400)
im = Image.open('lena.png')
region = im.crop(box)
plt.imshow(region)
```

## Processing a subrectangle, and pasting it back

- `im.transpose()`
- `Image.paste(im, box=None, mask=None)`

```python
region = region.transpose(Image.ROTATE_180)
plt.imshow(region)
```

```python
im.paste(region, box)
plt.imshow(im)
```

## Rolling an image

- `im.load()` if not called, the crop will not be done until the _paste_ is called

```python
def roll(img, delta):
    xsize, ysize = img.size
    delta = delta % xsize
    part1 = img.crop((0, 0, delta, ysize))
    part2 = img.crop((delta, 0, xsize, ysize))
    
    part1.load()
    part2.load()
    
    img.paste(part2, (0, 0, xsize - delta, ysize))
    img.paste(part1, (xsize - delta, 0, xsize, ysize))
    
    return img

plt.imshow(roll(Image.open('lena.png'), 256))
```

## Spliting and merging bands

```python
im = Image.open('lena.png')
r, g, b = im.split()

plt.subplot(221)
im = Image.merge("RGB", (b, g, r))
plt.imshow(im)

plt.subplot(222)
im = Image.merge("RGB", (r, g, b))
plt.imshow(im)

plt.subplot(223)
im = Image.merge('RGB', (b, r, g))
plt.imshow(im)

plt.subplot(224)
im = Image.merge('RGB', (r, b, g))
plt.imshow(im)
```

# Geometrical transforms

```python
im = Image.open('lena.png')

out = im.resize((128, 128))
plt.subplot(221)
plt.imshow(out)

plt.subplot(222)
out = im.rotate(45)
plt.imshow(out)
```

## Transposing an image

```python
plt.figure(figsize=(30, 30))
_, axes = plt.subplots(2, 3)
axes[0, 0].imshow(im.transpose(Image.FLIP_LEFT_RIGHT))
axes[0, 1].imshow(im.transpose(Image.FLIP_TOP_BOTTOM))
axes[0, 2].imshow(im.transpose(Image.ROTATE_90))
axes[1, 0].imshow(im.transpose(Image.ROTATE_180))
axes[1, 1].imshow(im.transpose(Image.ROTATE_270))
axes[1, 2].imshow(im)
```

# Color Transform

`im.convert(mode)`

When translating a color image to black and white (mode “L”), the library uses the ITU-R 601-2 luma transform:

`L = R * 299/1000 + G * 587/1000 + B * 114/1000`

```python
im = Image.open('lena.png').convert('L')
plt.imshow(im)
```
